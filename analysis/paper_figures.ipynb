# Transfer Pathways local figure workspace

Read canonical source or curated data, do the analysis in this notebook, and publish only the finished matplotlib figure. The server never runs notebook code.

```python
import matplotlib.pyplot as plt
import starter as pmt
```

## Read the complete catalogs

These are full catalogs for every included CC and UC campus, not a computer-science-only subset.

```python
cc_courses = pmt.get("exports/courses")
uc_courses = pmt.get("exports/university-courses")
len(cc_courses), len(uc_courses)
```

## Build a figure locally

```python
receivers = pmt.get("exports/receivers")
counts = receivers.groupby("school").size().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(9, 5))
counts.plot.bar(ax=ax)
ax.set_ylabel("receiving requirements")
ax.set_xlabel("")
fig.tight_layout()
fig
```

## Publish the finished files

```python
pmt.publish(
    fig,
    slug="requirements-by-campus",
    title="Requirements per UC campus",
)
```
